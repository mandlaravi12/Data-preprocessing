```python
seed=np.arange(0,9)
mean_1=[]
std_1=[]
for s in seed:
  np.random.seed(s)
  sample_2=df[df["quality"].isin([6,7])].sample(10)[["alcohol"]].reset_index().drop(columns=["index"])
  mean_1.append(np.mean(sample_2["alcohol"]))
  std_1.append(np.std(sample_2["alcohol"]))

sample_data=pd.DataFrame(columns=["mean_1","std_1"],data=zip(mean_1,std_1))
sample_data
```

```python
mean_of_data=np.mean(sample_data["mean_1"])
mean_of_data
```

```python
std_of_data=np.std(sample_data["mean_1"])
std_of_data
```

```python
standard_error=np.array(sample_data["mean_1"]).std()
standard_error
```

```python
sns.distplot(sample_data["mean_1"],hist=False)
```
